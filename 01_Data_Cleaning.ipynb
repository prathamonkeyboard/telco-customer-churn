# 1. Data Cleaning & Preparation

## Project Goal
Predict customer churn for a Telco company to improve retention.

## This Notebook
In this step, we prepare the raw dataset for analysis.
1. **Load Data**: Import the raw Telco Customer Churn dataset.
2. **Clean Data**: Handle missing values and fix data types (specifically `TotalCharges`).
3. **Export**: Save the clean version for the next step.

```python
import pandas as pd 
import numpy as np 

df = pd.read_csv('../data/raw/Telco-Customer-Churn.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df["TotalCharges"]= pd.to_numeric(df["TotalCharges"], errors="coerce")
df["TotalCharges"].isnull().sum()
```

```python
# removing null values
df.dropna(inplace=True)

# removing customerid as it's not important in analysis
df.drop(columns=["customerID"], inplace=True)
```

```python
df.head()
```

```python
print(df['Churn'].unique())
```

```python
#Changing the Churn column values: 1 for Yes, and 0 for No
df["Churn"]=df["Churn"].map({"Yes": 1, "No": 0})
```

```python
df.head()
```

```python
df.shape
```

## Output
Saving the clean dataset to `data/processed/telco_churn_cleaned.csv`.

```python
df.to_csv('../data/processed/telco_churn_cleaned.csv', index=False)

print("Clean data saved to data/processed/telco_churn_cleaned.csv")
```
